```python
#exact one histogram with and without built in function(in single code)
import cv2
import numpy as np
import matplotlib.pyplot as plt

def calculateHist(img):
    hist = np.zeros(256)
    rows, cols = img.shape[:2]
    for i in range(rows):
        for j in  range(cols):
            intensity = img[i,j]
            hist[intensity] += 1
    return hist

if img is None: 
    image = cv2.imread('cat1.jpg', cv2.IMREAD_GRAYSCALE)
else:
    
    hist1 = cv2.calcHist([image], [0], None, [256], [0, 256])

    plt.figure(figsize=(12, 8))

    plt.subplot(1, 3, 1)
    plt.imshow(image, cmap='gray')
    plt.title('Original Image')

    plt.subplot(1, 3, 2)
    plt.plot(hist1)
    plt.title('Histogram(using built-in)')

    hist2 = calculateHist(image)
    plt.subplot(1, 3, 3)
    plt.plot(hist2)
    plt.title('Histogram(without using built-in)')
```

```python
if sorted(hist1 )==sorted(hist2):
    print("same list")
else:
    print("different")
```

```python
#calculate average smoothing using built in function
import cv2
import numpy as np
import matplotlib.pyplot as plt


def average_smooth(img,kernel_size):
    rows,cols=img.shape
    modified_img=np.zeros((rows,cols))
    for i in range(rows):
        for j in range(cols):
            sum_1=0
            try:
                for p in range(-(kernel_size//2),kernel_size//2+1):
                    for q in range(-(kernel_size//2),kernel_size//2+1):
                        sum_1+=img[(i+p,j+q)]
            except IndexError as e:
                pass
            modified_img[i,j]=sum_1/(kernel_size**2)
    return modified_img


img=input("enter image::")

# Kernel size for smoothing
kernel_size = int(input("enter kernel size::"))

if img is None:
    print("Error:Unable to load the image")
else:
# Load the image in grayscale mode
    img = cv2.imread(img, cv2.IMREAD_GRAYSCALE)


# Apply average smoothing using cv2.blur
    smoothed_image1 = cv2.blur(img, (kernel_size, kernel_size))
    
# Display the original and smoothed images
    plt.figure(figsize=(12, 8))

# Original image
    plt.subplot(1, 3, 1)
    plt.title("Original Image")
    plt.imshow(img, cmap='gray')

# Smoothed image
    plt.subplot(1, 3, 2)
    plt.title("Smoothed Image(built in)")
    plt.imshow(smoothed_image1, cmap='gray')

    smooth_img2=average_smooth(img,kernel_size)

    plt.subplot(1, 3, 3)
    plt.title("Smoothed image(without built) ")
    plt.imshow(smooth_img2,cmap='gray')

# Show the plots
    plt.show()
```

```python
import numpy as np
if np.array_equal(smoothed_image1, smooth_img2):
    print("The smoothed images are the same.")
else:
    print("The smoothed images are not the same.")
```
