```python
#row major order 
import cv2
img=cv2.imread('sunflower.jpg',0)
print("original image=",img)
rows,cols=img.shape
rowvector=[]
for i in range(rows):
    for j in range(cols):
        rowvector.append(img[i,j])
print("row major order =",rowvector)
```

```python
# column major order
import cv2
img=cv2.imread('sunflower.jpg',0)
print("original image=",img)
rows,cols=img.shape
colvector=[]
for i in range(cols):
    for j in range(rows):
        colvector.append(img[j,i])
print("column major order=",colvector)
```

```python
#program to compute euclidean distance
import math
v1=[2,5,9,8,12]
v2=[13,16,22,24,25]
sum_1=0
for i in range(len(v1)):
    sum_1+=(v1[i]-v2[i])**2
ED=math.sqrt(sum_1)
print('Euclidean distance= ',ED)
```

```python
#Read image and compute average and variance(using built in)
import cv2
import numpy as np
def average_variance(img):
    avg=np.mean(img)
    variance=np.var(img)
    return avg,variance
img=cv2.imread('sunflower.jpg',0)
average,var=average_variance(img)
print("Average",average)
print("variance",var)
```

```python
#(exact question)Read image and compute average and variance(without using built in)
import cv2
def average_variance(img):
    rows,cols=img.shape
    total=0
    variance_sum=0
    for i in range(rows):
        for j in range(cols):
            total+=img[i,j]
    avg=total/(rows*cols)
    #variance
    for i in range(rows):
        for j in range(cols):
            variance_sum+=(img[i,j]-avg)**2
    variance=variance_sum/(rows*cols)
    return avg,variance
imgpath=input("Enter image path:")
img=cv2.imread(imgpath,cv2.IMREAD_GRAYSCALE)
# img=cv2.imread('sunflower.jpg',0)
print("original image",img)
average,variance=average_variance(img)
print("Average",average)
print("variance",variance)
```
